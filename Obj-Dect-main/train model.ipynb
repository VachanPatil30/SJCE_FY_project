```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, GlobalAveragePooling2D
from keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D
from keras.layers import BatchNormalization
from keras.models import Model

from keras.applications.mobilenet import MobileNet
```

```python
img_rows = 224
img_cols = 224

model = MobileNet(weights='imagenet'  , include_top = False ,input_shape=(img_rows, img_cols, 3))


for  layer in model.layers:
    layer.trainable = False
    
    

for (i,layer) in enumerate(model.layers):
    print(str(i) + " "+ layer.__class__.__name__, layer.trainable)
    
```

```python
top = model.output
top = GlobalAveragePooling2D()(top)
print(top)
top = Dense(1024, activation='relu')(top)
top = Dense(256, activation='relu')(top)
top = Dense(128, activation='relu')(top)
top = Dense(80, activation='softmax')(top)




for (i,layer) in enumerate(model.layers):
    print(str(i) + " "+ layer.__class__.__name__, layer.trainable)
    
```

```python
from keras.models import Model
model  = Model(inputs = model.input , outputs = top )


for (i,layer) in enumerate(model.layers):
    print(str(i) + " "+ layer.__class__.__name__, layer.trainable)
    
print(model.summary())
```

```python

from keras.preprocessing.image import ImageDataGenerator

train_data_dir = 'dataset/train/'
#validation_data_dir = 'users/ec2-user/coco/images/test/'
# set our batch size (typically on most mid tier systems we'll use 16-32)
batch_size = 32

train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=45,
      width_shift_range=0.3,
      height_shift_range=0.3,
      horizontal_flip=True,
      fill_mode='nearest')
 
#validation_datagen = ImageDataGenerator(rescale=1./255)
 
train_generator = train_datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_rows, img_cols),
        batch_size=batch_size,
        class_mode='categorical')
 
#validation_generator = validation_datagen.flow_from_directory(
#       validation_data_dir,
#       target_size=(img_rows, img_cols),
#       batch_size=batch_size,
#       class_mode='categorical')


from keras.optimizers import RMSprop
from keras.callbacks import ModelCheckpoint, EarlyStopping

                     
checkpoint = ModelCheckpoint("mygmobileNet.h5",
                             monitor="val_loss",
                             mode="min",
                             save_best_only = True,
                             verbose=1)

earlystop = EarlyStopping(monitor = 'val_loss', 
                          min_delta = 0, 
                          patience = 3,
                          verbose = 1,
                          restore_best_weights = True)

# we put our call backs into a callback list
callbacks = [earlystop, checkpoint]

# We use a very small learning rate 
model.compile(loss = 'categorical_crossentropy',
              optimizer = RMSprop(lr = 0.001),
              metrics = ['accuracy'])

# Enter the number of training and validation samples here
nb_train_samples = 1097
nb_validation_samples = 272

# We only train 5 EPOCHS 
epochs = 3
batch_size = 16

history = model.fit_generator(
    train_generator,
    steps_per_epoch = nb_train_samples // batch_size,
    epochs = epochs,
    callbacks = callbacks,
    #validation_data = validation_generator,
    #validation_steps = nb_validation_samples // batch_size
)
```

```python
model.save("mymobilenet.h5")
```
